```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from scipy.signal import firwin
from scipy.signal import freqz
from scipy.signal import find_peaks_cwt
from scipy.signal import butter, lfilter
import peakutils
from peakutils.plot import plot as pplot
from matplotlib import pyplot

%matplotlib inline

ordata = np.genfromtxt("stringContactResult.csv", dtype=float, delimiter=',', names=True)
```

```python
size = len(ordata["Param_size"])/3;
x = ordata["Param_size"][0:size]
score = ordata["Score"]
plt.plot(x, score[0:size], label = 'plus op')
plt.plot(x, score[size:size * 2], label = 'StringBuffer')
plt.plot(x, score[size*2:size *3], label = 'StringBuilder')
plt.xlabel('loop size')
plt.ylabel('average cost (ms)')
plt.legend()
```
